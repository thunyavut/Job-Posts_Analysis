```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('GD_Jobs.csv')
df.set_index('index', inplace=True)
df
```

### Check and drop duplicate rows

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

### Data cleansing

```python
fig, ax = plt.subplots(1,2)
plt.rcParams["figure.figsize"] = (15,5)
sns.histplot(x = "Rating", data = df, ax=ax[0])
sns.histplot(x = "Founded", data = df, ax=ax[1])
plt.show()
```

#### Deal with unwanted values

```python
df.Rating = np.where(df.Rating == -1.0, 0, df.Rating)
df.Founded = np.where(df.Founded == -1.0, '-', df.Founded)
```

```python
print(df[df['Rating'] == -1].shape)
print(df[df['Founded'] == -1].shape)
df
```

```python
df.replace('-1','Unknown',inplace = True)
```

```python
df[(df == '-1') | (df == -1)].any()
```

#### Take unwanted data out

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
df['Company Name'] = df['Company Name'].apply(lambda x: x.split('\n')[0])
df
```

### Extract skills from Description to show explicitly in dataframe

```python
df['C'] = df['Job Description'].apply(lambda x: 1 if 'c' in x.lower().replace('"',' ').replace("''",' ').replace(",",' ').split() else 0)
df['C++'] = df['Job Description'].apply(lambda x: 1 if 'c++' in x.lower() else 0)
df['C#'] = df['Job Description'].apply(lambda x: 1 if 'c#' in x.lower() else 0)
df['Java'] = df['Job Description'].apply(lambda x: 1 if 'java' in x.lower() else 0)
df['Python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['Excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['Hadoop'] = df['Job Description'].apply(lambda x: 1 if 'hadoop' in x.lower() else 0)
df['Spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['AWS'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['Tableau'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df['Power BI'] = df['Job Description'].apply(lambda x: 1 if 'power bi' in x.lower() else 0)
df['Big data'] = df['Job Description'].apply(lambda x: 1 if 'big data' in x.lower() else 0)
df
```

### Show estimate min/max salary

```python
df['Min Salary Estimate (thousand dollars)'] = df['Salary Estimate'].apply(lambda x: int(x.replace('K','').replace('$','').split('-')[0]))
df['Max Salary Estimate (thousand dollars)'] = df['Salary Estimate'].apply(lambda x: int(x.replace('K','').replace('$','').split('-')[1]))
```

```python
df.drop('Salary Estimate', axis=1, inplace=True)
```

### Show in-demand positon and seniority

```python
def Title(title):
    if 'data scientist' in title.lower():
        return 'Data scientist'
    elif 'data engineer' in title.lower() or 'data science' in title.lower():
        return 'Data engineer'
    elif 'computer scientist' in title.lower():
        return 'Computer Scientist'
    elif 'analyst' in title.lower():
        return 'Analyst'
    elif 'software engineer' in title.lower():
        return 'Software Engineer'
    elif 'machine learning' in title.lower():
        return 'Ml'
    elif 'manager' in title.lower():
        return 'Manager'
    elif 'director' in title.lower():
        return 'Director'
    else:
        return 'Please read a Title'
    
def seniority(title):
    if 'sr' in title.lower() or 'sr.' in title.lower() or 'senior' in title.lower() or 'leader' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'Senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower() or 'junior' in title.lower():
        return 'Junior'
    else:
        return 'Read a description'
```

```python
df['Seniority'] = df['Job Title'].apply(seniority)
df['Position'] = df['Job Title'].apply(Title)
```
